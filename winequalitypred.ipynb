```python
X= wine_data.drop(columns='quality')
X.head()
```

```python
Y = wine_data['quality'].apply(lambda y_value: 1 if y_value >= 7 else 0)
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(Y.shape, y_train.shape , y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
rc_classifier= RandomForestClassifier()
rc_classifier.fit(x_train,y_train)
```

```python
y_pred = rc_classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
y_pred_acc = accuracy_score(y_pred,y_test)
```

```python
print('Accuracy:',y_pred_acc)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(30,30))
for i in range(len(rc_classifier.estimators_)):
    plot_tree(rc_classifier.estimators_[i],filled=True)
```
